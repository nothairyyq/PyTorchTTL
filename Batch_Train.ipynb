## Batch Concept
对于一个有2000个训练样本的数据集，将2000个sample分成大小为500的batch，完成一个epoch需要4个iteration

在一个 epoch 中，batch 数和迭代数是相等的

### DataLoader

```python
torch.manual_seed(1) #reproducible
BATCH_SIZE=9  # Batch traing 的每一个batch的数据个数

x=torch.linspace(1,10,10)
y=torch.linspace(10,1,10)

#先转换成torch能识别的dataset
torch_dataset=Data.TensorDataset(x,y)

#把dataset放入dataloader
loader=Data.DataLoader(
    dataset=torch_dataset,   #shape:torch tensordataset format
    batch_size=BATCH_SIZE,
    shuffle=False,       #将数据打乱
    num_workers=2,      #多线程来读取数据
)
```

## Train

```python
for epoch in range(3):
    for step,(batch_x,batch_y) in enumerate(loader):
        #train
        #pass
         print('Epoch: ', epoch, '| Step: ', step, '| batch x: ',
              batch_x.numpy(), '| batch y: ', batch_y.numpy())
```
